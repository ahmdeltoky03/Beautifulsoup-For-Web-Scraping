```python
## import main Libraries
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url = 'https://books.toscrape.com/catalogue/category/books_1/index.html'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
## Book Title
title = soup.find('title')
title = title.get_text().strip().split(' | ')[0]
title
```

```python
## Type of Books That Exist in The Library and URL of It's Type
books = soup.find_all('ul')[2]
books = books.find_all('li')
type_of_book = []
urls = []
for book in books:
    # print('http://books.toscrape.com/catalogue/category'+ book.find('a').get('href')[2:])
    type_of_book.append(book.get_text().strip())
    urls.append('https://books.toscrape.com/catalogue/category'+book.find('a').get('href')[2:])

## Print Books_Type : Books_Type_Link
for type,url in zip(type_of_book,urls):
    print(type," : ",url)
```

```python
## Rate_Dictionary
rate_dic = {'One':1,'Two':2,'Three':3,'Four':4,"Five":5}
```

```python
books_names = [] ## Added init Books_Name in all type 
books_prices = [] ## Added init all Books_Price in all type 
books_rates = [] ## Added init all Books_Rate in all type
types = []
## loop in All_URL To append each List of Books in each URL
for _url,type in zip(urls,type_of_book):
    _response = requests.get(_url)
    _soup = BeautifulSoup(_response.text,'html.parser') 
    books = _soup.find_all('li',attrs = {'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
    

    book_name = [] ## Added init Books_Name in each type 
    book_price = [] ## Added init all Books_Price in each type 
    book_rate = [] ## Added init all Books_Rate in each type

    ## books_Name
    for book in books:
        book_name.append((book.find('h3').find('a').get('title')))
        # print(book.find('h3').find('a').get('title'))
        # print('-'*100)


    ## books_Price
    for book in books:
        book_price.append(book.find('p',attrs={'class':"price_color"}).get_text()[2:])
        # print(book.find('p',attrs={'class':"price_color"}).get_text())
        # print('-'*100)

    ## books_Rate
    for book in books:
        book_rate.append(book.find('p',attrs={'class':'star-rating'}).get('class')[-1])
        # books_rate.append(book.find('p').get('class')[-1])
        # print(book.find('p',attrs={'class':'star-rating'}).get('class')[-1])
        # print('-'*100

    for name,price,rate in zip(book_name,book_price,book_rate):
        books_names.append(name)
        books_prices.append(price)
        books_rates.append(rate)
        types.append(type)
    
```

```python
## create CSV file
_fieldnames = ['Books_Name','Books_Price','Books_Rate','Type']
with open ('Library.csv','w',newline='',encoding='utf-8') as f:
    writer = csv.DictWriter(f,fieldnames=_fieldnames)
    writer.writeheader()
    for name,price,rate,type in zip(books_names,books_prices,books_rates,types):
        writer.writerow({'Books_Name':name,
                         'Books_Price':price,
                         'Books_Rate':rate_dic[rate],
                            'Type':type})
```
