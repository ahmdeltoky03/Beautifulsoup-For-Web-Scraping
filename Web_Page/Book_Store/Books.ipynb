# `Books to Scrape`

## **URL** =  `http://books.toscrape.com/index.html`

This is a library in which a group of books is displayed,
and each group of books falls under a specific concept.
For example, books related to travel, mysteries, children...etc.
If you want to search for a specific type of book,
enter the URL of the subject at the top,
and then choose the type. What is required,
copy the URL and put it in the requested URL.
Once run all cells in Jupter is done,
the books are collected in a CSV file
with the name of the chosen type and kept on the device.


```python
## import main libraries
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
## Enter Your which  url you want to extract data from it
requried_url = 'https://books.toscrape.com/catalogue/category/books/fantasy_19/index.html'
response = requests.get(requried_url)
```

```python
response.content
# response.text
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
## Title of Any Backage of Books
_title = soup.find('title').get_text().strip().split(' | ')[0]
_title
```

```python
_title
```

```python
books = soup.find_all('li',attrs = {'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
books
```

```python
## books_Name
books_name = []
for book in books:
    books_name.append((book.find('h3').find('a').get('title')))
    # print(book.find('h3').find('a').get('title'))
    # print('-'*100)

## books_Price
books_price = []
for book in books:
    books_price.append(book.find('p',attrs={'class':"price_color"}).get_text()[2:])
    # print(book.find('p',attrs={'class':"price_color"}).get_text())
    # print('-'*100)

## books_Rate
books_rate = []
for book in books:
    books_rate.append(book.find('p',attrs={'class':'star-rating'}).get('class')[-1])
    # books_rate.append(book.find('p').get('class')[-1])
    # print(book.find('p',attrs={'class':'star-rating'}).get('class')[-1])
    # print('-'*100)

## create CSV file
_fieldnames = ['Books_Name','Books_Price','Books_Rate']
with open (f'{_title}.csv','w',newline='',encoding='utf-8') as f:
    writer = csv.DictWriter(f,fieldnames=_fieldnames)
    writer.writeheader()
    for name,price,rate in zip(books_name,books_price,books_rate):
        writer.writerow({'Books_Name':name,
                         'Books_Price':price,
                         'Books_Rate':rate})
## Read CSV file using Pandas
book_store = pd.read_csv(f'{_title}.csv')
```

```python
book_store
```

```python
## Average price of the book (Mean)
# Average = sum(books_price)/len(books_price)
summation_of_price = 0
for price in books_price:
    summation_of_price += float(price)
summation_of_price/len(books_price) ## mean
```
