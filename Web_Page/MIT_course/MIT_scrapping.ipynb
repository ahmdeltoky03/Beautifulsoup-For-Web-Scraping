import requests
from bs4 import BeautifulSoup

```python
url = 'https://ocw.mit.edu/courses/6-006-introduction-to-algorithms-spring-2020/video_galleries/lecture-videos/'
response = requests.get(url)
response.text
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
data = soup.find('div',attrs = {'class':'max-content-width course-banner-content m-auto py-3'})
```

```python
data
```

```python
## course name
file_name = data.find('a').get_text()
```

```python
## coures name
# soup.find('title').get_text().split(' | ')[1]
soup.find('title').get_text().split('|')[1].strip()
```

```python
## lectures
videos = soup.find_all('div',attrs = {'class':'mb-2 border-gray rounded video-gallery-card'})
videos
```

```python
## lecture_name and lecture_link
for vid in videos:
    lec_url = 'https://ocw.mit.edu' +vid.find('a').get('href'))
    lec_name = vid.find('h5').get_text()
```

```python
import csv
fieldnames = ["Lecture_Names","Lecture_Link"]
with open ("Introduction to Algorithms.csv","w",newline='',encoding='utf-8') as f:
    writer = csv.DictWriter(f,fieldnames=fieldnames)
    writer.writeheader()
    for vid in videos:
        lec_url = 'https://ocw.mit.edu' +vid.find('a').get('href')
        lec_name = vid.find('h5').get_text()
        writer.writerow({'Lecture_Names':lec_name,'Lecture_Link':lec_url})
        
```
